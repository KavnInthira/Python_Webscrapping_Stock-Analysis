### Calculating 100 day moving average for the Adjusted Close

```python
close_FSR=df_FSR['Adj Close']
df_FSR['100ma'] = close_FSR.rolling(window=100,min_periods = 0).mean()
sma_FSR=df_FSR['100ma']
df_FSR.head()
```

```python
close_GM=df_GM['Adj Close']
df_GM['100ma'] = close_GM.rolling(window=100,min_periods = 0).mean()
sma_GM=df_GM['100ma']
df_GM.head()
```

```python
close_LCID=df_LCID['Adj Close']
df_LCID['100ma'] = close_LCID.rolling(window=100,min_periods = 0).mean()
sma_LCID=df_LCID['100ma']
df_LCID.head()
```

### Candle chart with volume

```python
df_ohlc_FSR = df_FSR['Adj Close'].resample('10D').ohlc()
df_volume_FSR = df_FSR['Volume']. resample('10D').sum()

df_ohlc_FSR.reset_index(inplace=True)
df_ohlc_FSR['Date'] =df_ohlc_FSR['Date'].map(mdates.date2num)

ax1 =plt.subplot2grid((6,1), (0,0), rowspan=4, colspan=1)
plt.title('Candle Chart for FSR')
ax2 =plt.subplot2grid((6,1), (5,0), rowspan=1, colspan=1, sharex=ax1)
plt.title('Volume for FSR')
ax1.xaxis_date()

candlestick_ohlc(ax1, df_ohlc_FSR.values, width=2, colorup='g')
ax2.fill_between(df_volume_FSR.index.map(mdates.date2num),df_volume_FSR.values,0)
plt.show()
```

```python
df_ohlc_GM = df_GM['Adj Close'].resample('10D').ohlc()
df_volume_GM = df_GM['Volume']. resample('10D').sum()

df_ohlc_GM.reset_index(inplace=True)
df_ohlc_GM['Date'] =df_ohlc_GM['Date'].map(mdates.date2num)

ax1 =plt.subplot2grid((6,1), (0,0), rowspan=4, colspan=1)
plt.title('Candle Chart for GM')
ax2 =plt.subplot2grid((6,1), (5,0), rowspan=1, colspan=1, sharex=ax1)
plt.title('Volume for GM')
ax1.xaxis_date()

candlestick_ohlc(ax1, df_ohlc_GM.values, width=2, colorup='g')
ax2.fill_between(df_volume_GM.index.map(mdates.date2num),df_volume_GM.values,0)
plt.show()
```
